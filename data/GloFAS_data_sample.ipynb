```python
ax1 = df.plot.scatter(x=10,y=11,c='Blue')
dfdyn.plot.scatter(x=10,y=11,c='Green',ax=ax1)
ax1.set_ylabel('Latitude')
ax1.set_xlabel('Longitude')
ax1.set_title('Static and Dynamic Reporting Points')
plt.show()
```
